```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# داده‌ها
data = {
    'Model': ['XGBoost', 'ANN', 'SVM'],
    'MAE_Top7': [0.18, 1.24, 0.76],
    'MAE_Top10': [1.44, 1.66, 1.56],
    'RMSE_Top7': [0.26, 2.65, 0.92],
    'RMSE_Top10': [1.88, 4.19, 3.89]
}

df = pd.DataFrame(data)

# تنظیمات نمودار
x = np.arange(len(df['Model']))  # موقعیت مدل‌ها
width = 0.35  # عرض میله‌ها

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# نمودار MAE
axs[0].bar(x - width/2, df['MAE_Top7'], width, label='Top 7 Features')
axs[0].bar(x + width/2, df['MAE_Top10'], width, label='Top 10 Features')
axs[0].set_title('MAE Comparison')
axs[0].set_xticks(x)
axs[0].set_xticklabels(df['Model'])
axs[0].set_ylabel('MAE')
axs[0].legend()

# نمودار RMSE
axs[1].bar(x - width/2, df['RMSE_Top7'], width, label='Top 7 Features')
axs[1].bar(x + width/2, df['RMSE_Top10'], width, label='Top 10 Features')
axs[1].set_title('RMSE Comparison')
axs[1].set_xticks(x)
axs[1].set_xticklabels(df['Model'])
axs[1].set_ylabel('RMSE')
axs[1].legend()

plt.tight_layout()
plt.savefig('comparison_chart.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# داده‌ها
data = {
    'Model': ['ANN', 'ANN', 'SVM', 'SVM', 'XGBoost', 'XGBoost'],
    'Feature Set': ['Top 7', 'Top 10', 'Top 7', 'Top 10', 'Top 7', 'Top 10'],
    'Mean MAE': [1.62, 1.7142, 0.9791, 1.7342, 0.2154, 1.7296],
    'Mean MSE': [4.35, 4.4826, 1.6084, 4.4826, 0.0901, 4.4591]
}

df = pd.DataFrame(data)

# ایجاد دو نمودار در یک صفحه
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# نمودار MAE
for feature_set in ['Top 7', 'Top 10']:
    subset = df[df['Feature Set'] == feature_set]
    axes[0].bar(subset['Model'] + ' (' + feature_set + ')', subset['Mean MAE'], label=f'MAE - {feature_set}')
axes[0].set_title('Mean MAE Comparison')
axes[0].set_ylabel('Mean MAE')
axes[0].tick_params(axis='x', rotation=45)
axes[0].legend()

# نمودار MSE
for feature_set in ['Top 7', 'Top 10']:
    subset = df[df['Feature Set'] == feature_set]
    axes[1].bar(subset['Model'] + ' (' + feature_set + ')', subset['Mean MSE'], label=f'MSE - {feature_set}')
axes[1].set_title('Mean MSE Comparison')
axes[1].set_ylabel('Mean MSE')
axes[1].tick_params(axis='x', rotation=45)
axes[1].legend()

plt.tight_layout()
plt.savefig('comparison_chart.jpg')
plt.show()
```
